```python
import matplotlib.pyplot as plt
import seaborn as sns

# Polarity dağılımı için histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Polarity'], bins=30, kde=True, color='skyblue')
plt.title('Polarity (Duygu) Dağılımı')
plt.xlabel('Polarity (Duygu)')
plt.ylabel('Frekans')
plt.grid(True)
plt.show()

# Subjectivity dağılımı için histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Subjectivity'], bins=30, kde=True, color='lightgreen')
plt.title('Subjectivity (Öznelik) Dağılımı')
plt.xlabel('Subjectivity (Öznelik)')
plt.ylabel('Frekans')
plt.grid(True)
plt.show()

# Polarity ve Subjectivity arasındaki ilişkiyi gösteren bir scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Polarity', y='Subjectivity', color='orange')
plt.title('Polarity ve Subjectivity Arasındaki İlişki')
plt.xlabel('Polarity (Duygu)')
plt.ylabel('Subjectivity (Öznelik)')
plt.grid(True)
plt.show()
```
