```python
import pandas as pd
import numpy as np

# Rastgele veri oluşturma
np.random.seed(42)

# İtalyan isimleri
italian_names = [
    "Luca", "Giulia", "Marco", "Francesca", "Matteo", "Chiara", "Alessandro", "Sara", "Davide", "Martina",
    "Andrea", "Elena", "Francesco", "Valentina", "Simone", "Alice", "Giorgio", "Federica", "Lorenzo", "Silvia"
]

# Yemekler (İtalyan yemekleri)
meals = ["Pizza Margherita", "Spaghetti Carbonara", "Risotto alla Milanese", "Lasagna", "Tiramisu", 
         "Focaccia", "Gnocchi", "Panna Cotta", "Bruschetta", "Cannoli"]

# Restoranlar (İtalyan restoran isimleri)
restaurants = ["Ristorante Roma", "Trattoria Milano", "Osteria Firenze", "Pizzeria Napoli", "Cucina Siciliana"]

# Sipariş geçmişi
orders = []
for _ in range(1000):
    order = {
        "user_name": np.random.choice(italian_names),
        "meal": np.random.choice(meals),
        "restaurant": np.random.choice(restaurants),
        "rating": np.random.randint(1, 6),  # 1 ile 5 arasında rastgele bir puan
        "order_date": pd.to_datetime(f"2023-{np.random.randint(1, 13)}-{np.random.randint(1, 29)}")
    }
    orders.append(order)

# DataFrame oluşturma
orders_df = pd.DataFrame(orders)

# CSV dosyasına kaydetme
orders_df.to_csv("italian_orders_with_names.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# En popüler yemekler
popular_meals = orders_df['meal'].value_counts().head(5)

# En popüler restoranlar
popular_restaurants = orders_df['restaurant'].value_counts().head(5)

# Grafik oluşturma
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
popular_meals.plot(kind='bar', color='#800000')  # Canlı bordo
plt.title('En Popüler Yemekler')
plt.xlabel('Yemek')
plt.ylabel('Sipariş Sayısı')

plt.subplot(1, 2, 2)
popular_restaurants.plot(kind='bar', color='#00008B')  # Koyu lacivert
plt.title('En Popüler Restoranlar')
plt.xlabel('Restoran')
plt.ylabel('Sipariş Sayısı')

plt.tight_layout()
plt.show()
```

```python
# Kullanıcıların ortalama puanları
user_ratings = orders_df.groupby('user_name')['rating'].mean().sort_values(ascending=False).head(10)

# Grafik oluşturma
plt.figure(figsize=(8, 5))
user_ratings.plot(kind='bar', color='#556B2F')  # Canlı turuncu
plt.title('En Yüksek Ortalama Puan Veren Kullanıcılar')
plt.xlabel('Kullanıcı İsmi')
plt.ylabel('Ortalama Puan')
plt.show()
```

```python
import pandas as pd

# CSV dosyasını okuma
orders_df = pd.read_csv('italian_orders_with_names.csv')

# En popüler yemekler ve tatlılar
popular_meals_and_desserts = orders_df['meal'].value_counts()

# Yüzde hesaplama
total_orders = popular_meals_and_desserts.sum()
popular_meals_and_desserts_percentage = (popular_meals_and_desserts / total_orders) * 100

print("En Popüler Yemekler ve Tatlılar Yüzdeleri:")
print(popular_meals_and_desserts_percentage)

# En popüler restoranlar
popular_restaurants = orders_df['restaurant'].value_counts()

# Yüzde hesaplama
total_restaurant_orders = popular_restaurants.sum()
popular_restaurants_percentage = (popular_restaurants / total_restaurant_orders) * 100

print("\nEn Popüler Restoranlar Yüzdeleri:")
print(popular_restaurants_percentage)
```

```python
import pandas as pd

# CSV dosyasını okuma
orders_df = pd.read_csv('italian_orders_with_names.csv')

# Kullanıcıların ortalama puanlarını hesaplama
user_ratings = orders_df.groupby('user_name')['rating'].mean().sort_values(ascending=False)

# En yüksek ortalama puan veren kullanıcılar
top_user_ratings = user_ratings.head(10)

# Yüzde hesaplama
total_users = len(user_ratings)
top_user_ratings_percentage = (top_user_ratings / total_users) * 100

print("En Yüksek Ortalama Puan Veren Kullanıcılar Yüzdeleri:")
print(top_user_ratings_percentage)
```
